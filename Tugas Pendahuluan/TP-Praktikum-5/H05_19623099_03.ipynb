Mengimport Pandas dan Matplotlib, serta membuat dataframe dengan nama 'df' yang membaca "H05.csv"

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("H05.csv")
```

1. Histogram distribusi nilai Matematika.

```python
df["nilai_mat"].plot(kind = "hist", xticks = [i for i in range(0, 101, 5)], yticks = [i for i in range(0, 1001, 50)], range = [0, 100], bins = [i for i in range(0, 101, 5)], rwidth = 1, xlabel = "Nilai Matematika", ylabel = "Banyak mahasiswa", title = "Histogram Distribusi Nilai Matematika", edgecolor = "black")
plt.show()
```

2. Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas.

```python
df["fakultas"].value_counts().plot(kind = "barh", xticks = [i for i in range(0, 1001, 100)], xlabel="Banyak Mahasiswa", ylabel="Fakultas", title = "Diagram Batang Horizontal Banyak Mahasiswa per Fakultas", edgecolor = "black")
plt.show()
```

3. Diagram pie banyaknya mahasiswa masing-masing fakultas.

```python
df["fakultas"].value_counts().plot(kind = "pie", autopct = "%1.1f%%", ylabel = "", title = "Diagram Pie Banyak Mahasiswa per Fakultas")
plt.show()
```

4. Berdasarkan diagram pie dan batang horizontal, manakah fakultas dengan mahasiswa terbanyak? Manakah diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak dan mengapa?

    Berdasarkan diagram pie dan batang horizontal, fakultas dengan mahasiswa terbanyak adalah FTI. Dari antara dua diagram ini, diagram yang lebih baik dalam menampilkan data adalah diagram batang horizontal karena tampilan data menunjukkan secara lebih akurat banyak mahasiswa tiap fakultas, sementara diagram pie tidak secara eksplisit menunjukkan banyak mahasiswa melainkan hanya persentasenya dalam diagram.

```python
df.plot(kind = "scatter", title = "Scatter Plot Nilai Kimia dengan Nilai Fisika", x = "nilai_kim", y = "nilai_fis", xlabel = "Nilai Kimia", ylabel = "Nilai Fisika", xticks = [i for i in range(0, 101, 10)], yticks = [i for i in range(0, 101, 10)])
plt.show()
```
