```python
import pandas as pd
df = pd.read_csv('P05-03.csv')
```

```python
print(len(df.loc[(df['minimum_nights'] == 1) & (df['number_of_reviews'] != 1)]))
```

```python
availabilitymax = df['availability_365'].max()
print(df['host_name'].loc[df['availability_365'] == availabilitymax].to_string(index = False))
```

```python
print(len(df.loc[((df['neighbourhood_group'] == 'Central Region') | (df['neighbourhood_group'] == 'East Region')) & (df['price'] < 100)]))
```

```python
print(df["room_type"].loc[df['reviews_per_month'] > 1].value_counts().to_string())
```

```python
nhostmax = df['host_name'].value_counts().idxmax()
hostmax = df.loc[df['host_name'] == nhostmax, 'name'].value_counts()
print(hostmax.head(5).to_string())
```
