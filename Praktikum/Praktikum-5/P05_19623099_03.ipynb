```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('P05-03.csv')
```

```python
dfprivateroom = df.loc[df['room_type'] == 'Private room']
dfprivateroom['number_of_reviews'].plot(kind="hist", bins = [i for i in range(0, 301 , 30)], rwidth =0.8, xticks = [i for i in range(0, 301, 50)])
plt.show
None
```

```python
dfreview100 = df.loc[df['number_of_reviews'] > 100]
dfreview100['room_type'].value_counts().plot(kind = 'pie')
plt.show
None
```

```python
dfnotjay = df.loc[df['host_name'] != 'Jay']
dfnotjay.plot(kind="scatter", x="longitude", y="latitude")
plt.show
None
```

```python
dfnovena = df.loc[df['neighbourhood'] == 'Novena']
dfnovena.plot(kind="scatter",x="longitude", y="latitude",sizes=df["availability_365"])
plt.show
None
```

```python
dfjej = df.loc[(df['host_name'] == 'JoJo') | (df['host_name'] == 'Elizabeth') | (df['host_name'] == 'Jimmy')]
dfjej.groupby(["room_type","host_name"])["name"]. size ().unstack ().plot(kind="bar", stacked=True)
plt.show
None
```
